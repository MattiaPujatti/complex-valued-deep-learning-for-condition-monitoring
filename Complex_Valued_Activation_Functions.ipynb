# Complex-Valued Activation Functions

```python
# Dependencies
import numpy as np
import matplotlib.pyplot as plt
import cplot
from mpl_toolkits.mplot3d import axes3d
from matplotlib import cm

%matplotlib inline
```

### Complex Sigmoid

$$
f(z) = \frac{1}{1+e^{-z}}
$$

```python
def complex_sigmoid(z):
    return 1. / (1. + np.exp(-z))
```

3D and contour plot:
